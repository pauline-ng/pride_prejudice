```python
import networkx as nx  #load networkx package

# create a graph
Graph = nx.Graph()

# add the relationships to the graph
# each person is a node, relation specifies relationship
Graph.add_edge('Mr. Bingley','Mr. Darcy',relation ='BFFâ€™')
Graph.add_edge ("Mr. Darcy", "Lady de Bourgh", relation="aunt / nephew")
Graph.add_edge ("Lizzie Lim", "Jane Lim", relation="sisters")
Graph.add_edge ("Margaret", "Lou", relation="just married!")
Graph.add_edge ("Margaret", "Lizzie Lim", relation="cousin")
Graph.add_edge ("Margaret", "Jane Lim", relation="cousin")
Graph.add_edge ("Lou", "Mr. Bingley", relation="coworker")

pos = nx.spring_layout (Graph)

#visualize network
nx.draw (Graph, pos,  with_labels=True)

#add edge labels

nx.draw (Graph, pos)
```

```python
nx.draw_networkx_edge_labels(Graph, pos, edge_labels=nx.get_edge_attributes(Graph, 'relation'))
```

```python

# distance or degrees of separation between Mr. Darcy and lizzie
p = nx.shortest_path_length (Graph, source="Lizzie Lim", target="Mr. Darcy")
print (p)

Graph.add_edge ("Jane Lim", "Mr. Bingley", relation="possibly dating")
p = nx.shortest_path_length (Graph, source="Lizzie Lim", target="Mr. Darcy")
print (p)
```
