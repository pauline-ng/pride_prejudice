```python
import networkx as nx  #load networkx package
import json
from networkx.readwrite import json_graph
import os

def read_graph_from_json_file(filename):
    with open(filename) as f:
        js_graph = json.load(f)
    return json_graph.node_link_graph(js_graph)

def download_network ():
    import urllib.reques
    url = "https://raw.githubusercontent.com/pauline-ng/pride_prejudice/main/network.json"
    urllib.request.urlretrieve (url, 'network.json')
    
    
# check that the file exists, otherwise download it
if not os.path.exists ('network.json'):
    download_network()
    
Graph = read_graph_from_json_file ('network.json')
```

```python
# draw graph, to make sure we read it in
pos = nx.spring_layout (Graph)
nx.draw (Graph, pos, with_labels=True)
```

```python

# find the distance or "degrees of separation between" Mr. Darcy and lizzie
p = nx.shortest_path_length (Graph, source="Lizzie Lim", target="Mr. Darcy")

print ("Degrees of separation between Lizzie and Mr.Darcy is : " + str(p))

```
